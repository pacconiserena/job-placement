```python
import numpy
import pandas
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

Leggo i file generati da PREPARATION con il metodo read_csv

```python
file_train = pandas.read_csv("train.csv")
file_test = pandas.read_csv("test.csv")
```

```python
# con dtypes vedo il tipo di dato di ogni colonna
file_train.dtypes
```

```python

labelEncoder = LabelEncoder()

file_test['gender']=labelEncoder.fit_transform(file_test['gender'])
file_train['gender']=labelEncoder.fit_transform(file_train['gender'])

file_test['ssc_p']=labelEncoder.fit_transform(file_test['ssc_p'])
file_train['ssc_p']=labelEncoder.fit_transform(file_train['ssc_p'])

file_test['ssc_b']=labelEncoder.fit_transform(file_test['ssc_b'])
file_train['ssc_b']=labelEncoder.fit_transform(file_train['ssc_b'])

file_test['hsc_p']=labelEncoder.fit_transform(file_test['hsc_p'])
file_train['hsc_p']=labelEncoder.fit_transform(file_train['hsc_p'])

file_test['hsc_b']=labelEncoder.fit_transform(file_test['hsc_b'])
file_train['hsc_b']=labelEncoder.fit_transform(file_train['hsc_b'])

file_test['hsc_s']=labelEncoder.fit_transform(file_test['hsc_s'])
file_train['hsc_s']=labelEncoder.fit_transform(file_train['hsc_s'])

file_test['degree_p']=labelEncoder.fit_transform(file_test['degree_p'])
file_train['degree_p']=labelEncoder.fit_transform(file_train['degree_p'])

file_test['degree_t']=labelEncoder.fit_transform(file_test['degree_t'])
file_train['degree_t']=labelEncoder.fit_transform(file_train['degree_t'])

file_test['workex']=labelEncoder.fit_transform(file_test['workex'])
file_train['workex']=labelEncoder.fit_transform(file_train['workex'])

file_test['etest_p']=labelEncoder.fit_transform(file_test['etest_p'])
file_train['etest_p']=labelEncoder.fit_transform(file_train['etest_p'])

file_test['specialisation']=labelEncoder.fit_transform(file_test['specialisation'])
file_train['specialisation']=labelEncoder.fit_transform(file_train['specialisation'])

file_test['mba_p']=labelEncoder.fit_transform(file_test['mba_p'])
file_train['mba_p']=labelEncoder.fit_transform(file_train['mba_p'])

file_test['salary']=labelEncoder.fit_transform(file_test['salary'])
file_train['salary']=labelEncoder.fit_transform(file_train['salary'])

file_test['status']=labelEncoder.fit_transform(file_test['status'])
file_train['status']=labelEncoder.fit_transform(file_train['status'])

file_train.dtypes
```

```python
#utilizzo set_option per vedere tutte le righe
pandas.set_option("max_rows", None)

file_train.head(140)
```

```python
#rimuovo colonna salary da entrambi i file perchè sono presenti valori nulli
file_train = file_train.drop(["salary"], axis=1) 
file_test = file_test.drop(["salary"], axis=1)

#stampo la tabella per vedere se è stata eliminata la colonna salary
file_train.head(140)
```

```python
#Rimuovo dalla tabella la colonna status per calcolare l'accuracy

status_train = file_train['status']
train_withoutStatus = file_train.drop(['status'], axis=1)

status_test = file_test['status']
test_withoutStatus = file_test.drop(['status'], axis=1)
```

Calcolo l'accuratezza del dataset

```python
classifier = RandomForestClassifier()
classifier.fit(train_withoutStatus, status_train)
status_predict = classifier.predict(test_withoutStatus)
accuracy = round(accuracy_score(status_predict, status_test) * 100, 2)
print(accuracy)
```
